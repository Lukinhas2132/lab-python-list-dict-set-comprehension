````python
def initialize_inventory(products):
    """
    Initializes the inventory by prompting the user for the quantity of each product.
    Uses a dictionary comprehension to build the inventory.
    """
    inventory = {
        product: int(input(f"Enter the quantity of {product}s available: "))
        for product in products
    }
    return inventory

def get_customer_orders():
    """
    Prompts the user for the number of orders and collects the requested products.
    Uses a set comprehension to store unique product orders.
    """
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {
        input("Enter the name of a product that a customer wants to order: ")
        for _ in range(num_orders)
    }
    return customer_orders

def update_inventory(customer_orders, inventory):
    """
    Decrements the inventory for each ordered product (if it exists).
    Removes products that have 0 quantity remaining.
    Utilizes dictionary comprehension to filter out zero-quantity items.
    """
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
    
    # Filter out products with zero quantity
    inventory = {p: q for p, q in inventory.items() if q > 0}
    return inventory

def calculate_order_statistics(customer_orders, products):
    """
    Calculates:
      1. Total number of unique products ordered.
      2. Percentage of the total products (out of available product list) that were ordered.
    Returns both values as a tuple.
    """
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered

def print_order_statistics(order_statistics):
    """
    Prints the total products ordered and the percentage of unique products ordered.
    """
    print("\nOrder Statistics:")
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Unique Products Ordered:", order_statistics[1])

def print_updated_inventory(inventory):
    """
    Prints the updated inventory after fulfilling the customer orders.
    """
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

def calculate_total_price(customer_orders):
    """
    Prompts the user to enter a price for each product in the order (assumes quantity=1).
    Uses a list comprehension to collect prices, then sums them.
    Returns the total price of the order.
    """
    prices = [
        float(input(f"Enter the price of {product}: "))
        for product in customer_orders
    ]
    return sum(prices)

# Main program execution

if __name__ == "__main__":
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    
    # 1. Initialize the inventory
    inventory = initialize_inventory(products)
    
    # 2. Get the customer's orders
    customer_orders = get_customer_orders()
    
    # 3. Update the inventory
    inventory = update_inventory(customer_orders, inventory)
    
    # 4. Calculate and print order statistics
    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)
    
    # 5. Print the updated inventory
    print_updated_inventory(inventory)
    
    # 6. Calculate the total price of the order and print it
    total_price = calculate_total_price(customer_orders)
    print("Total Price:", total_price)
````
